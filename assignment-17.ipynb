```python
"""
Matplotlib is a plotting library for the Python programming language and its numerical mathematics
extension, NumPy. It provides an object-oriented API for embedding plots into applications using
general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK.

Matplotlib is used for creating static, animated, and interactive visualizations in Python. 
It is widely used for data visualization and exploration, as well as for creating publication-quality 
figures in various formats.

examples are bar plot, scatterplot, histogram, scatter plot, box plot
"""
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
"""
The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure.
It allows you to arrange plots in a grid-like layout, specifying the number of rows, columns, and the 
position of each plot within the grid.
"""
import matplotlib.pyplot as plt
import numpy as np


data = [
    (np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([50, 20, 40, 20, 60, 70])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60])),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150]))
]


fig, axs = plt.subplots(2, 2, figsize=(12, 8))


for i, ax in enumerate(axs.flat):
    x, y = data[i]
    ax.plot(x, y)
    ax.set_title(f'Line Plot {i+1}')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
"""
A bar plot is a type of plot that presents categorical data with rectangular bars, where the length of each 
bar represents the value of the data. 
Bar plots are used to compare different categories of data or to show changes in data over time.
"""
```

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(10, 5))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Profit of Companies')
plt.show()

# Horizontal bar plot
plt.figure(figsize=(10, 5))
plt.barh(company, profit, color='salmon')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Profit of Companies')
plt.show()
```

```python
"""

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of 
a dataset based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), 
and maximum. It is used to visualize the spread and skewness of the data, as well as to identify outliers.
"""

import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.figure(figsize=(10, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot of Box 1 and Box 2')
plt.grid(True)
plt.show()
```
