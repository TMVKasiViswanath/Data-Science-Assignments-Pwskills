```python

Probability Mass Function (PMF): This is used for discrete random variables. It gives the probability that
a discrete random variable is exactly equal to some value. Mathematically, for a discrete random variable
X, the PMF is defined as P(X = x), where x is a specific value that X can take.

For example, consider a fair six-sided die. The PMF of rolling the die is:

P(X = 1) = 1/6
P(X = 2) = 1/6
P(X = 3) = 1/6
P(X = 4) = 1/6
P(X = 5) = 1/6
P(X = 6) = 1/6

Here, the PMF gives the probability of each outcome (rolling a 1, 2, 3, 4, 5, or 6) of the die.

Probability Density Function (PDF): This is used for continuous random variables. It gives the probability 
that a continuous random variable falls within a particular range of values. Mathematically, for a 
continuous random variable X, the PDF is denoted as f(x) and represents the derivative of the cumulative 
distribution function (CDF) of X.

For example, consider a continuous random variable X that represents the height of students in a class. 
The PDF of X might be a bell-shaped curve, such as the normal distribution. The PDF gives the probability
density at each point along the x-axis, but to find the probability of X falling within a range, you would
need to integrate the PDF over that range.

The integral of the PDF over the entire range of possible values is equal to 1, indicating that the total
probability of all possible outcomes is 1.
```

```python

The Cumulative Density Function (CDF) is a function used in probability theory and statistics to describe 
the probability that a random variable X will take on a value less than or equal to a specific value x.
Mathematically, for a random variable X, the CDF is defined as F(x) = P(X ≤ x).

The CDF provides a way to understand the distribution of a random variable by showing how probabilities
are spread across its possible values. It is particularly useful for determining the likelihood of a random
variable falling within a certain range of values.

Example:

Consider a fair six-sided die. The CDF of rolling the die is:

F(1) = P(X ≤ 1) = 1/6
F(2) = P(X ≤ 2) = 2/6
F(3) = P(X ≤ 3) = 3/6
F(4) = P(X ≤ 4) = 4/6
F(5) = P(X ≤ 5) = 5/6
F(6) = P(X ≤ 6) = 6/6 = 1
In this example, the CDF shows the cumulative probabilities of rolling a value less than or equal to each 
possible outcome of the die. For instance, the probability of rolling a value less than or equal to 3 is 
3/6, or 0.5.

Why CDF is used:

Provides a comprehensive view: The CDF gives a complete picture of the distribution of a random variable by
showing the probabilities for all possible values.

Easy comparison: It allows for easy comparison of probabilities across different values of the random 
variable.

Calculating probabilities: The CDF can be used to calculate the probability of a random variable falling 
within a specific range of values by subtracting the probabilities at the endpoints of the range.

Determining percentiles: It can also be used to determine percentiles, such as the median (50th percentile) 
or quartiles (25th and 75th percentiles), which are important measures of central tendency and spread in a 
distribution.

```

```python

The uniform distribution is a probability distribution where all outcomes are equally likely. 
In other words, each value in the distribution has the same probability of occurring. It is often
represented as a rectangular-shaped distribution, where the height of the rectangle represents the
probability density.

Consider a fair six-sided die. When the die is rolled, each face (1, 2, 3, 4, 5, or 6) has an equal 
probability of 1/6 of being rolled. This is an example of a discrete uniform distribution.

For a continuous uniform distribution, consider a random variable X representing the time it takes for
a bus to arrive at a bus stop, where buses are scheduled to arrive every 10 minutes. The arrival time of
the bus is uniformly distributed between 0 and 10 minutes, meaning that any minute within that interval
is equally likely for the bus to arrive.
```

```python

The z-score, also known as the standard score, is a measure of how many standard deviations a data point 
is away from the mean of a dataset. It is calculated using the formula:
z-score=x-mu/sigma

Standardization: The z-score allows for the standardization of different datasets, making it easier to 
compare data from different distributions or populations.

Outlier Detection: Z-scores can be used to identify outliers in a dataset. Data points with z-scores that
are significantly higher or lower than the mean may be considered outliers and warrant further investigation.

Probability Calculation: In statistics, the z-score is used to calculate probabilities for normal 
distributions. The standard normal distribution table, also known as the z-table, provides the cumulative
probabilities for different z-scores, which can be used to calculate probabilities for any normal 
distribution.

Quality Control: In quality control processes, z-scores can be used to determine whether a process 
is within acceptable limits. Data points with z-scores outside certain thresholds may indicate a process
that is out of control and in need of adjustment.
```

```python
The Central Limit Theorem (CLT) is a fundamental theorem in probability theory and statistics. 
It states that the distribution of the sum (or average) of a large number of independent, identically
distributed random variables approaches a normal distribution, regardless of the shape of the original
distribution.

Approximation of Distributions: The Central Limit Theorem provides a powerful tool for approximating the 
distribution of sample means from any population, regardless of its original distribution. This allows 
statisticians to make inferences about population parameters based on sample data.

Basis for Hypothesis Testing: Many statistical tests and procedures, such as t-tests and confidence intervals, rely on the Central Limit Theorem. It enables us to use the normal distribution as an approximation for the sampling distribution of the sample mean, even when the population distribution is not normal.

Real-world Applications: The Central Limit Theorem is applicable in various fields, such as economics, biology, and engineering, where data are often collected from samples. It allows researchers to make statistical inferences and draw conclusions about populations based on sample data.

Quality Control: In manufacturing and quality control, the Central Limit Theorem is used to analyze and monitor processes. It helps in understanding the variation in product quality and making decisions to improve processes.
```

```python
Independence: The random variables being averaged or summed in the sample must be 
independent of each other. This means that the outcome of one variable should not affect the outcome of 
another.

Identically Distributed: The random variables should be identically distributed. This means that they 
should have the same probability distribution, including the same mean and variance.

Finite Variance: The random variables should have a finite variance. If the variance is infinite, the CLT 
may not hold.

Sample Size: The sample size should be sufficiently large. While there is no strict rule for how large the 
sample size should be, a common guideline is that the sample size should be greater than or equal to 30 
for the CLT to provide a good approximation.
```
