```python
"""
Scatter Plot:

Use: To visualize the relationship between two numerical variables.
Function: sns.scatterplot(x='x_variable', y='y_variable', data=df)
Line Plot:

Use: To show trends over a continuous interval or time.
Function: sns.lineplot(x='x_variable', y='y_variable', data=df)
Bar Plot:

Use: To compare categories of data.
Function: sns.barplot(x='x_variable', y='y_variable', data=df)
Histogram:

Use: To visualize the distribution of a single numerical variable.
Function: sns.histplot(x='variable', data=df)
Box Plot:

Use: To show the distribution of a numerical variable across different categories.
Function: sns.boxplot(x='x_variable', y='y_variable', data=df)
"""
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


fmri_data = sns.load_dataset("fmri")

# Create a line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
plt.title('Signal vs. Timepoint for different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.legend(title='Event')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


titanic_data = sns.load_dataset("titanic")

# Create two box plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Age distribution across passenger classes')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Fare distribution across passenger classes')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


diamonds_data = sns.load_dataset("diamonds")

# Plot a histogram with 'cut' as hue
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack', bins=30)
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.legend(title='Cut')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


iris_data = sns.load_dataset("iris")

# Create a pair plot with hue for 'species'
sns.pairplot(iris_data, hue='species')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")


flights_pivot = flights_data.pivot("month", "year", "passengers")


# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Number of Passengers (Flights Dataset)')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
